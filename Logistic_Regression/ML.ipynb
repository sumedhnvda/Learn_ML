```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=42)
```

```python
X 
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
```

```python
logistic.fit(X_train,Y_train)
```

```python
y_pred=logistic.predict(X_test)
```

```python
logistic.predict_proba(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
s=accuracy_score(Y_test,y_pred)
c=confusion_matrix(Y_test,y_pred)
clsr=classification_report(Y_test,y_pred)
```

```python
print(s,c,clsr)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
penalty=["l1", "l2", "elasticnet"]
c_values=[100, 10, 1.0, 0.1, 0.01]
solver=["lbfgs", "liblinear", "newton-cg", "newton-cholesky","sag", "saga"]
```

```python
param=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
model=GridSearchCV(estimator=logistic,param_grid=param,scoring="accuracy",cv=5,n_jobs=-1)
```

```python
model
```

```python
model.fit(X_train,Y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
model=RandomizedSearchCV(estimator=logistic,param_distributions=param,scoring="accuracy",cv=5,n_jobs=-1)
```

```python
model.fit(X_train,Y_train)
```

```python
model.best_params_
```
